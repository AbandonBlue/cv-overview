## 1. 前言(問題定義)

Metric learning 就是透過一些訓練機制，使得相似`輸入`的 embedding 盡可能相似。

好處在於相似輸入擁有相似embedding，可以下游任務的搜尋、推薦、比對等任務的優化。

```python
import random
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from collections import defaultdict
from PIL import Image
from sklearn.metrics import ConfusionMatrixDisplay
from tensorflow import keras
from tensorflow.keras import layers
```

## 2. 取得資料
使用公開 cifar-10。

```python
from tensorflow.keras.datasets import cifar10

(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# 標準化(加速收斂)

x_train = x_train / 255.0
x_test = x_test / 255.0
```

**簡單透過pil查看圖片資料，對資料有更清晰的理解**

```python
height_width = 32


def show_collage(examples):
    """
        透過PIL套件，創造一個圖(拼貼畫)，用樣本真實數值回填到這個圖上，使得一張圖上可以顯示多張圖的概念。
        用plt.subplots、imshow也可以達到類似概念。
    """
    global height_width
    box_size = height_width + 2
    num_rows, num_cols = examples.shape[:2]

    collage = Image.new(
        mode="RGB",
        size=(num_cols * box_size, num_rows * box_size),
        color=(250, 250, 250),
    )
    for row_idx in range(num_rows):
        for col_idx in range(num_cols):
            array = (np.array(examples[row_idx, col_idx]) * 255).astype(np.uint8)
            collage.paste(
                Image.fromarray(array), (col_idx * box_size, row_idx * box_size)
            )

    # Double size for visualisation.
    collage = collage.resize((2 * num_cols * box_size, 2 * num_rows * box_size))
    return collage
```

```python
# Show a collage of 5x5 random images.
sample_idxs = np.random.randint(0, 50000, size=(5, 5))
examples = x_train[sample_idxs]

print(examples.shape)

show_collage(examples)
```

## 3. 處理資料

`Metric Learning`需要的訓練資料與一般的(X, y)訓練資料不同，而是用多個彼此相似的樣本(instances)。在這次的example中，使用的是同一個類別代表`相似`，但在其他情況下，可以用自行定義的`相似`來建造訓練資料。所以單一個訓練樣本不是一張圖片，而是一對類別相同的`(相似)`圖片。在Metric Learning中，我們將`anchor`作為基礎圖片，`positive`則是相似的圖片(同類別)。

為了要方便取得樣本，用dict去建立class to indices。

```python
y_train
```

```python
## 建立mapping

class_idx_to_train_idxs = defaultdict(list)
class_idx_to_test_idxs = defaultdict(list)

for idx, class_ in enumerate(y_train):
    class_idx_to_train_idxs[class_[0]].append(idx)

for idx, class_ in enumerate(y_test):
    class_idx_to_test_idxs[class_[0]].append(idx)

# 類別0: 前10筆
class_idx_to_train_idxs[0][:10]
```

最簡單的產生訓練樣本的方法是，一個batch包含了(anchor, positive)以及其他類別而組成。訓練的目標是同一個batch之中的anchor、positive的向量在向量空間中盡量靠近，以其他類別盡量遠離。

```python
num_classes = 10

class AnchorPositivePairs(keras.utils.Sequence):
    """
        可參考: https://www.tensorflow.org/api_docs/python/tf/keras/utils/Sequence
        需要override `__getitem__`、`__len__`
    """
    def __init__(self, num_batchs, class_idx_to_train_idxs):
        self.num_batchs = num_batchs
        self.class_idx_to_train_idxs = class_idx_to_train_idxs
    
    def __len__(self):
        return self.num_batchs
    
    def __getitem__(self, idx):
        x = np.empty(shape=(2, num_classes, height_width, height_width, 3), dtype=np.float32) # 相當於20張圖片
        for class_idx in range(num_classes):
            # 每一個類別
            examples_for_class = self.class_idx_to_train_idxs[class_idx] # 取得class_idx類別的index樣本(全部)
            anchor_idx = random.choice(examples_for_class)               # 隨機選一張當作anchor圖片
            positive_idx = random.choice(examples_for_class)             # 再隨機選一張同類別當做positive圖片
            while positive_idx == anchor_idx: # 避免兩張是同一張圖片
                positive_idx = random.choice(examples_for_class)
            x[0, class_idx] = x_train[anchor_idx]    # 位置(0, class_idx)的地方放置anchor圖案
            x[1, class_idx] = x_train[positive_idx]  # Q: 其他位置呢? 不是只有 np.empty的數值嗎?不用用其他類別替代? Ans: 是一個迴圈，所以其他類別也會對應產生，所以這20張圖片是上下對應的！
            
            ## 嘗試看看把其他位置補成其他類別??
            
        return x
```

## 4. 模型

透過 `custom model`的`train_step`來實踐細節，也就是上述資料的`anchor`、`positive`的`Embedding`來進行相似計算，產出logits來輸出。

```python
tf.range(10)
```

```python
class EmbeddingModel(keras.Model):
    def train_step(self, data):
        if isinstance(data, tuple):
            data = data[0]
        anchors, positives = data[0], data[1]  # (10, 32, 32, 3), (10, 32, 32, 3) 上下對應
        
        with tf.GradientTape() as tape:
            anchor_embeddings = self(anchors, training=True)    # 進入模型取得embedding, (10, 8)
            positive_embeddings = self(positives, training=True) # (10, 8)
            #print(anchor_embeddings.shape, positive_embeddings.shape) # (batch_size, 8), (batch_size, 8)
            
            # 這是一種方便的向量計算方式，
            similarities = tf.einsum(
                "ae, pe->ap", anchor_embeddings, positive_embeddings
            )  # 這個的動作是在說，10張圖片互相 dot-product，相當於計算相似性
            #print(similarities.shape) # (None, None) = (batch_size, batch_size) 
            
            temperature = 0.2  # 這個我知道有這個東西，但具體理論不清楚
            #similarities /= temperature
            
            sparse_labels = tf.range(num_classes)  # 這邊怎麼弄的不應該是(batch_size,)嗎 ---> 因為其實是(10=num_classes, 10=num_classes)的類別，且只有主對角線上是正確的，所以這樣沒問題是in-batch訓練技巧
            loss = self.compiled_loss(sparse_labels, similarities)  # 因為是sparse_labels所以剛好也只有10個是對的預測值，沒有問題！
        
        gradients = tape.gradient(loss, self.trainable_variables)
        self.optimizer.apply_gradients(zip(gradients, self.trainable_variables))
        
        self.compiled_metrics.update_state(sparse_labels, similarities)
        return {m.name: m.result() for m in self.metrics}
```

接著建立將`image data`轉換成embedding的神經網路。

```python
inputs = keras.Input(shape=(height_width, height_width, 3))
x = layers.Conv2D(filters=32, kernel_size=3, strides=2, activation='relu')(inputs)
x = layers.Conv2D(filters=64, kernel_size=3, strides=2, activation='relu')(x)
x = layers.Conv2D(filters=128, kernel_size=3, strides=2, activation='relu')(x)
x = layers.GlobalMaxPool2D()(x)
embeddings = layers.Dense(units=8, activation=None)(x)
#embeddings = tf.nn.l2_normalize(embeddings, axis=-1)  # 沿著 axis 標準化數值, 不理解為什麼需要? 因為這樣搭配 tf.einsum 才是cosine similarity

model = EmbeddingModel(inputs, embeddings)
model.summary()
```

```python
next(AnchorPositivePairs(num_batchs=1000, class_idx_to_train_idxs=class_idx_to_train_idxs))
```

```python
## 訓練

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['acc']
)

history = model.fit(AnchorPositivePairs(num_batchs=1000, class_idx_to_train_idxs=class_idx_to_train_idxs), epochs=20)

plt.plot(history.history["loss"])
plt.show()
```

```python
near_neighbours_per_example = 10

embeddings = model.predict(x_test)  # 因為這是 predict ，不會再predict 那邊有 tf.range的問題
gram_matrix = np.einsum("ae,be->ab", embeddings, embeddings)
near_neighbours = np.argsort(gram_matrix.T)[:, -(near_neighbours_per_example + 1) :]
```

```python
## 一定是方形，對角線都是正樣本

tf.einsum(
    "ae, pe->ap", embeddings, embeddings
)
```

```python
embeddings.shape
```

```python
num_collage_examples = 5

examples = np.empty(
    (
        num_collage_examples,
        near_neighbours_per_example + 1,
        height_width,
        height_width,
        3,
    ),
    dtype=np.float32,
)
for row_idx in range(num_collage_examples):
    examples[row_idx, 0] = x_test[row_idx]
    anchor_near_neighbours = reversed(near_neighbours[row_idx][:-1])
    for col_idx, nn_idx in enumerate(anchor_near_neighbours):
        examples[row_idx, col_idx + 1] = x_test[nn_idx]

show_collage(examples)
```

```python
confusion_matrix = np.zeros((num_classes, num_classes))

# For each class.
for class_idx in range(num_classes):
    # Consider 10 examples.
    example_idxs = class_idx_to_test_idxs[class_idx][:10]
    for y_test_idx in example_idxs:
        # And count the classes of its near neighbours.
        for nn_idx in near_neighbours[y_test_idx][:-1]:
            nn_class_idx = y_test[nn_idx]
            confusion_matrix[class_idx, nn_class_idx] += 1

# Display a confusion matrix.
labels = [
    "Airplane",
    "Automobile",
    "Bird",
    "Cat",
    "Deer",
    "Dog",
    "Frog",
    "Horse",
    "Ship",
    "Truck",
]
disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=labels)
disp.plot(include_values=True, cmap="viridis", ax=None, xticks_rotation="vertical")
plt.show()
```
